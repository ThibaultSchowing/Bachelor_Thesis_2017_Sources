```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
import os  
from openpyxl import load_workbook
import pandas as pd



def get_sheet_by_filename(filename):
    
    wb = load_workbook(filename, data_only=True, read_only=True)
    
    sheet_ind=0
    sheet_names = wb.get_sheet_names()
    sheet1 = wb.get_sheet_by_name(sheet_names[sheet_ind])
    
    return sheet1

def get_datas(ws,fn):
    # The order is similar to the csv parsed from pdfs. The empty chaines ("") are info not present in those files

    nombreCaficultor = ws['B6'].value
    # a ajouter au pdf
    cedula = ws['G6'].value
    #
    municipio = ""
    vereda = ws['B8'].value
    finca = ws['B7'].value
    sica = ""

    # à supprimer du pdf
    grupo = ""

    # à ajouter au pdf
    variedad = ws['G8'].value

    fechaRecpt = ""
    fechaAnalisis = ws['G7'].value

    # à supprimer du pdf
    noOrden = ""

    malla18 = ""
    malla17 = ""
    malla16 = ""
    malla15 = ""
    malla14 = ""
    malla13 = ""
    malla12 = ""
    malla0 = ""

    # Defectos fisicos solo %
    negrosYVinagres = ws['H13'].value

    #broca = brocaSevera ? 
    broca = ws['H14'].value
    brocaDePunto =ws['H15'].value

    veteado = ws['H16'].value
    mordido = ""
    inmaduro = ws['H18'].value
    flojo = ws['H19'].value
    sobresecado = ws['H20'].value
    arrugado = ws['H21'].value
    aplastado = ws['H22'].value
    cristalizado = ws['H23'].value
    reposado = ws['H25'].value
    granizo = ws['H26'].value
    conchas = ""
    partido = ws['H24'].value
    ambrar = ""
    defectosTotales = ws['H27'].value

    quarker = ""
    fluorescentes = ""

    # analisis fisico - dans l'ordre du premier csv !

    #on s'en fiche
    #volumenMuestra = vs[''].value

    Pergamino = ws['B15'].value
    Almendra = ""
    # Ajouter dans pdf
    AlmendraTotal = ws['B16'].value
    AlmendraSana = ws['B17'].value
    #

    UV = ""
    Olor = ws['B18'].value
    Humedad = ws['B14'].value
    Merma = ws['B19'].value
    FactorDeRendimiento = ws['B21'].value

    # ElementesExtranos on s'en fiche

    # analisis sensorial

    AromaFragancia = ws['C31'].value
    Acidez = ws['C34'].value
    Cuerpo = ws['C35'].value
    Sabor = ws['C32'].value
    SaborResidual = ws['C33'].value
    Dulzor = ws['C39'].value
    Uniformidad = ws['C36'].value
    Balance = ws['C37'].value
    TazaLimpia = ws['C38'].value
    PuntajeCatador = ws['C40'].value

    Taza1 = ws['G31'].value
    Taza2 = ws['G32'].value
    Taza3 = ws['G33'].value
    Taza4 = ws['G34'].value
    Taza5 = ws['G35'].value

    # A supprimer du pdf: Grupo , no de orden

    # A ajouter au pdf: cédula, variedad, Almendra total et sana + long et lat

    Longitude = ""
    Latitude = ""

    # Todo: vérifier pertinence "almendra" etc

    data = [nombreCaficultor, cedula, municipio, vereda, finca, sica, Longitude, Latitude, variedad, fechaRecpt, fechaAnalisis, 
            malla18, malla17, malla16, malla15, malla14, malla13, malla12, malla0,
            negrosYVinagres, broca, brocaDePunto, veteado, mordido, inmaduro, flojo, sobresecado,
            arrugado, aplastado, cristalizado, reposado, granizo, conchas, partido, ambrar, defectosTotales, 
            quarker, fluorescentes, Pergamino, Almendra, AlmendraTotal, AlmendraSana, UV, Olor, Humedad, Merma,
            FactorDeRendimiento, AromaFragancia, Acidez, Cuerpo, Sabor, SaborResidual, Dulzor, Uniformidad,
            Balance, TazaLimpia, PuntajeCatador, Taza1, Taza2, Taza3, Taza4, Taza5]
    return data





valuesTaza = []

for fn in os.listdir('./rawdatas'):
    ws = get_sheet_by_filename('./rawdatas/' + fn)
    data = get_datas(ws,fn)
    valuesTaza.append(data)

items = ['Nombre Caficultor', 'cedula', 'Municipio', 'Vereda', 'Finca', 'SICA','Longitude','Latitude', 'Variedad', 'Fecha Recpt', 'Fecha Analysis', 
         'Malla 18', 'Malla 17', 'Malla 16', 'Malla 15', 'Malla 14', 'Malla 13', 'Malla 12', 'Malla 0', 
         'Negros y vinagres', 'Broca', 'Broca de punto', 'Veteado', 'Mordido', 'Inmaduro', 'Flojo', 'Sobresecado',
         'Arrugado', 'Aplastado', 'Cristalizado', 'Reposado', 'Granizo', 'Conchas','Partido', 'Ambar', 'Defectos totales',
         'Quaker (No Granos)', 'Fluorescentes (No Granos)', 'Pergamino', 'Almendra','AlmendraTotal', 'AlmendraSana', 'UV', 'Olor', 'Humedad (%)', 'Merma (%)', 
         'Factor de Rendimiento', 'Aroma-Fragancia', 'Acidez', 'Cuerpo', 'Sabor', 'Sabor residual', 'Dulzor', 'Uniformidad', 
         'Balance', 'Taza limpia', 'Puntaje catador', 'Taza 1', 'Taza 2', 'Taza 3', 'Taza 4', 'Taza 5']

df = pd.DataFrame.from_records(valuesTaza, columns=items)
df
```
