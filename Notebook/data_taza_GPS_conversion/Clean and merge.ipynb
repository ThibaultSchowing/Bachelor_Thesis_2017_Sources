```python
data_noNaN2 = data_noNaN.reset_index(drop=True)

print data_noNaN2.index


inProj = Proj(init='epsg:3116')
outProj = Proj(init='epsg:4326')

data_noNaN2['EPSG:3116_Y'] = ""
data_noNaN2['EPSG:3116_X'] = ""

for index, row in data_noNaN2.iterrows():
    x1,y1 = row['CENT_X'],row['CENT_Y']
    #print x1, y1
    x2,y2 = transform(inProj,outProj,x1,y1)
    #print x2,y2
    data_noNaN2.set_value(index, 'EPSG:3116_X', x2)
    data_noNaN2.set_value(index, 'EPSG:3116_Y', y2)
    
    #data_noNaN[index]['EPSG:3116_X'] = x2
    #data_noNaN[index]['EPSG:3116_Y'] = y2


for index, row in data_noNaN2.iterrows():
    print row['EPSG:3116_Y'], row['EPSG:3116_X']
```
