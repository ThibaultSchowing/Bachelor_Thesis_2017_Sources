```python
import pandas as pd
import Averages as avgs

wrapper = avgs.ClimaticDataWraper(["2011","2012","2013","2014","2015","2016"])
```

```python
df = pd.DataFrame.from_csv("./DataWithGPS_v1_utf-8.csv", header=0, sep=',', index_col=0, parse_dates=False, encoding='utf-8')
df
```

```python
from datetime import *
#data_noNaN['Fecha Analysis'] = pd.to_datetime(data_noNaN['Fecha Analysis'])

for index, row in df.iterrows():
    try:
        date = datetime.strptime(row['Fecha Analysis'], '%m/%d/%Y %H:%M')
    except:
        date = datetime.strptime(row['Fecha Analysis'], '%m/%d/%Y')
    if date.year == 2017:
        date = date.replace(year = 2016)
    df.set_value(index,'Fecha Analysis', date)

# TODO Eliminate 2010
```

```python
# on va prendre les donn√©es climatiques pour 6 mois, jusqu'√† 1 mois avant la d√©gustation 
import numpy as np
def getClimatDatas(x,y,year,monthDegust):
    
    
    
    precdata = wrapper.getPrecAverages([x,y])
    tmindata = wrapper.getTminAverages([x,y])
    tmaxdata = wrapper.getTmaxAverages([x,y])
    tmeandata = wrapper.getTmeanAverages([x,y])
    dtrdata = wrapper.getDtrAverages([x,y])
    
    # Si data 2010 - remplacer
    focus = year - 2011
    
    
    m6 = (month - 2)
    m5 = (month - 3)
    m4 = (month - 4) 
    m3 = (month - 5) 
    m2 = (month - 6) 
    m1 = (month - 7) 
    monthsidx = np.array([m1,m2,m3,m4,m5,m6])
    
    # idx of the month in the data returned by the wrapper (0 = jan)
    '''
        focus:          0               1               2               3               4               5
                |      2011     |      2012     |      2013     |      2014     |      2015     |      2016     | 
                |Jan - ... - dec|Jan - ... - dec|Jan - ... - dec|Jan - ... - dec|Jan - ... - dec|Jan - ... - dec|
                
        Exemple:        xxxxxx    
        
                    7 d√©c 2011 => focus = 0 
                    idxs = 0 * 12 + [5,6,7,8,9,10]
                
        
    
    '''
    
    
    idxs = monthsidx + (focus * 12)
    
    # for one position we have 5 tables of datas 
    datastables = [precdata[idxs], tmindata[idxs], tmaxdata[idxs], tmeandata[idxs], dtrdata[idxs]]
    formated = []
    
    for val in precdata[idxs]: 
        formated.append(val)
    for val in tmindata[idxs]: 
        formated.append(val)
    for val in tmaxdata[idxs]: 
        formated.append(val)
    for val in tmeandata[idxs]: 
        formated.append(val)
    for val in dtrdata[idxs]: 
        formated.append(val)
    
    return formated

# Receive all climatic data before beeing merged to the dataframe
tmpdatas = []

for index, row in df.iterrows():
    x = row['EPSG:3116_X']
    y = row['EPSG:3116_Y']
    year = row['Fecha Analysis'].year
    month = row['Fecha Analysis'].month
    tmpdatas.append(getClimatDatas(x,y,year,month))
#
```

```python
tmpdatas = np.array(tmpdatas)
print tmpdatas.shape
```
